```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./images/sdocu.jpg')
img = img[..., ::-1] #BGR -> RGB 리버스

kernel_gx = np.array([[-1, 1]])
kernel_gy = np.array([[-1], [1]])

edge_gx = cv2.filter2D(img, -1, kernel_gx)
edge_gy = cv2.filter2D(img, -1, kernel_gy)
#x축 optical y축 horizontal로 검출
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, dpi=100)
ax1.imshow(img)
ax1.axis('off')
ax2.imshow(edge_gx)
ax2.axis('off')
ax3.imshow(edge_gy)
ax3.axis('off')
fig.tight_layout()
plt.show()
```

```python
import numpy as np
from skimage import io, data
from scipy import ndimage

coins = data.coins()
diff2d = np.array([
    [0,1,0],[1,0,-1],[0,-1,0]
])
coin_edges = ndimage.convolve(coins, diff2d)
io.imshow(coin_edges)
plt.show()
```
